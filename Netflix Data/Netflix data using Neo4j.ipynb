```python

word_list=[]
def cleanDes(desc): 
    import string
    tokens = word_tokenize(desc)
    # remove all tokens that are not alphabetic
    words = [word for word in tokens if word.isalpha()]
    # convert to lower case
    tokens = [w.lower() for w in tokens]
    # remove punctuation from each word
    table = str.maketrans('', '', string.punctuation)
    stripped = [w.translate(table) for w in tokens]
    # remove remaining tokens that are not alphabetic
    words = [word for word in stripped if word.isalpha()]
    # filter out stop words
    from nltk.corpus import stopwords
    stop_words = set(stopwords.words('english'))
    words = [w for w in words if not w in stop_words]
    # stemming of words
    lemma = nltk.wordnet.WordNetLemmatizer()
    stemmed = [lemma.lemmatize(word) for word in words]
    word_list.extend(stemmed)
    return stemmed


df.description=[cleanDes(desc) for desc in df.description]

final_word_list=[word for word,count in Counter(word_list).most_common()]
print(final_word_list)
```

```python
# Deleting existing nodes if any
def delNode():
    request="""
    MATCH (n) DETACH DELETE n;
    """
    print(graph.run(request))
```

```python
def addNode(nodeType,**kwargs):
    parameters=''
    for key, value in kwargs.items():
        if parameters=='':
            parameters='''%s:"%s"'''% (key, value)
        else:    
            parameters=parameters+","+'''%s:"%s"'''% (key, value)
    request ='''CREATE ({}:{}'''.format(nodeType.lower()[0],nodeType) + '''{'''+parameters +'''})'''
    graph.run(request)
```

```python
def createRelationMovie(title,MatchNode2,relation,node,to=True):
    match='''match (m:Movie {title:"'''+'''{}"'''.format(title)+'''}'''+'''),({}:{}'''.format(MatchNode2.lower()[0],MatchNode2)+''' {name:'''+'''"{}"'''.format(node)+'''})'''
    if to:
        create=''' create (m)-[:{}]->({})'''.format(relation,MatchNode2.lower()[0])
    else:
        create=''' create (m)<-[:{}]-({})'''.format(relation,MatchNode2.lower()[0])
    ret=''' return m,{}'''.format(MatchNode2.lower()[0])
    request=match+' '+create
    graph.run(request)
    
```

```python
def createRelationActors(name,MatchNode2,relation,node):
    match='''match (a:Person {name:"'''+'''{}"'''.format(name)+'''}'''+'''),({}:{}'''.format(MatchNode2.lower()[0],MatchNode2)+''' {name:'''+'''"{}"'''.format(node)+'''})'''
    create=''' create (a)-[:{}]->({})'''.format(relation,MatchNode2.lower()[0])
    request=match+' '+create
    graph.run(request)
```

```python

bar = progressbar.ProgressBar(maxval=len(words), \
    widgets=[progressbar.Bar('=', '[', ']'), ' ', progressbar.Percentage()])
bar_count=0
bar.start()

start_time_overall = time.time()
print("Adding Word nodes")

for i in df.iterrows():
    title=i[1]['title'].replace('"','')
    countries=i[1]['country'].split(",")
    countries=[name.strip() for name in countries if name.strip()]
    for country in countries:
        createRelationMovie(title,'Country','made_at',country)
    createRelationMovie(title,'Year','created_in',int(i[1]['year']))
    createRelationMovie(title,'Year','released_in',int(i[1]['release_year']))
    createRelationMovie(title,'Rating','has_rating',i[1]['rating'])  
    createRelationMovie(title,'Rating','has_type',i[1]['type']) 
    categories=i[1]['listed_in'].split(",")
    categories=[name.strip() for name in categories if name.strip()]
    for category in categories:
        createRelationMovie(title,'Category','listed_in',category)
    cast=i[1]['cast'].split(",")
    cast=[name.strip() for name in cast if name.strip()]
    for person in cast:
        createRelationMovie(title,'Person','acted',person,to=False)    
    for actor1,actor2 in set((itertools.combinations(cast, 2))):
        createRelationActors(actor1,'Person','works_with',actor2)
    director=i[1]['director'].split(",")
    director=[name.strip() for name in director if name.strip()]
    for person in director:
        createRelationMovie(title,'Person','directed',person,to=False)
    if i[1]['description']:
        words_description=[name.strip() for name in i[1]['description'] if name.strip()]
        for w in words_description:
            createRelationMovie(title,'Word','desc_contains',w)
    
    bar.update(bar_count+1)
    bar_count+=1
    
bar.finish()
exec_time_overall = time.time() - start_time_overall
print(f'-- {exec_time_overall} seconds --')
    
```
